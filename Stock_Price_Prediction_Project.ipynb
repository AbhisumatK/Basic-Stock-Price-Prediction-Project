# Target --> Whether stock will go UP or DOWN

```python
sp500['Tomorrow'] = sp500['Close'].shift(-1)
```

```python
sp500.tail()
```

```python
sp500['Target'] = (sp500['Tomorrow'] > sp500['Close']).astype(int)
```

```python
sp500 = sp500.loc['1990-01-01 00:00:00-05:00':].copy()
```

```python
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100, min_samples_split=25, random_state=42, class_weight='balanced')
```

```python
train = sp500.iloc[:-100]
test = sp500.iloc[-100:]
```

```python
predictors = [
    'Close',
    'Open',
    'High',
    'Low',
    'Volume'
]
```

# Base Model

```python
model.fit(train[predictors], train['Target'])
```

```python
from sklearn.metrics import precision_score
```

```python
pred = model.predict(test[predictors])
```

```python
preds = pd.Series(pred, index = test.index)
```

```python
def _predict(train, test, predictors, model):
    model.fit(train[predictors], train['Target'])
    pred = model.predict(test[predictors])
    preds = pd.Series(pred, index = test.index, name='Predictions')
    combined = pd.concat([test['Target'], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_preds = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()

        predictions = _predict(train, test, predictors, model)
        all_preds.append(predictions)

    return pd.concat(all_preds)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions['Predictions'].value_counts()
```

```python
precision_score(predictions['Target'], predictions['Predictions'])
```

```python
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()

    ratio_col = f'Close_Ratio_{horizon}'
    sp500[ratio_col] = sp500['Close'] / rolling_averages['Close']

    trend_col = f'Trend_{horizon}'
    sp500[trend_col] = sp500.shift(1).rolling(horizon).sum()['Target']

    new_predictors += [ratio_col, trend_col]
```
